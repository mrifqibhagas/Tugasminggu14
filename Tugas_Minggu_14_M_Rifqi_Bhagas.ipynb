```
import numpy as np
import matplotlib.pyplot as plt
```

```
ntl = lambda x: 10*np.log10(x)
ltn = lambda x: 10**(x/10)

def cos(frek,tipe,ht,hr,d):
    if frek in range(150,1501):
        Ca=69.55
        Cb=26.16
    elif frek in range(1501,2001):
        Ca=46.3
        Cb=33.9
    
    if tipe=='urban' :
        Cm=0
    elif tipe=='suburban' :
        Cm= -2*((np.log10((frek/28)**2))-5.4)
    elif tipe=='open' :
        Cm= -4.78*(np.log10(frek))+18.33*(np.log10(frek))-40.94
    else :
        Cm='tipe salah'
    
    ahr = (1.1*np.log10(frek)-0.7)*hr-(1.56*np.log10(frek)-1.8)
    Lp = Ca+(Cb*np.log10(frek))-(13.83*np.log(ht))-ahr+((44.9-6.55*np.log10(ht))*np.log10(d))-Cm
    
    return Lp
```

```
ht = 80
hr = 1.5
dmin=300 
dmax=1000 
N = 200
ptrns = 30
gtrns = 10
grcv = 2
bw = 5

d=np.random.randint(dmin,dmax,N)
```

```
def pin(ptrns,pl,gtrns,grcv):
  ptrns1 = ntl(ptrns)
  a = ptrns - pl + gtrns + grcv
  a1 = ltn(a)
  return a1
```

```
def snr(pin,bw):
  b = -174 + 10*np.log10(bw)
  b1=ltn(b)
  snr = pin/b1
  return snr
```

```
def ks(snr,bw):
  b = bw*2*np.log(1+snr)
  return ks
```

```

def pinurban():
  f=1800 
  loc='urban'
  pl=cos(f,loc,ht,hr,d)
  pinfix = pin(ptrns,pl,gtrns,grcv)
  d_sort=d[np.argsort(d)]
  Urban_sort=pinfix[np.argsort(d)]
  return d_sort,Urban_sort

def pinopen():
  f=1800 
  loc='open'
  pl=cos(f,loc,ht,hr,d)
  pinfix = pin(ptrns,pl,gtrns,grcv)
  d_sort=d[np.argsort(d)]
  Urban_sort=pinfix[np.argsort(d)]
  return d_sort,Urban_sort

def pinsuburban():
  f=1800 
  loc='suburban'
  pl=cos(f,loc,ht,hr,d)
  pinfix = pin(ptrns,pl,gtrns,grcv)
  d_sort=d[np.argsort(d)]
  Urban_sort=pinfix[np.argsort(d)]
  return d_sort,Urban_sort
```
